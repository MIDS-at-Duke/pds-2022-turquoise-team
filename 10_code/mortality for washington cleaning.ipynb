```python
# Load the dataset
vital = pd.read_table("../00_source_data/mortality for washington.txt")
vital
```

```python
# Drop the unnecessary data
vital = vital.drop(["Notes","Year Code"], axis = 1)
vital.sample(5)
```

```python
# Checking the amount of null values
vital.isna().sum()
```

```python
# Checking initial shape (before droppping null values)
vital.shape
```

```python
# Drop null values
vital.dropna(inplace=True)
```

```python
# Checking the shape after dropping null values
vital.shape
```

```python
# Adding state code 
vital["State_Code"] = vital["County"].str[-2:]
vital["County Name"] = vital["County"].str[:-4]
vital.sort_values("County")
```

```python
vital.sort_values("County")
```

```python
vital["Year"] = pd.DatetimeIndex(pd.to_datetime(vital["Year"],format="%Y")).year
vital
```

```python
vital["Cause of death"].value_counts()
```

```python
delete_cause = [
        "Mental and behavioural disorders due to use of tobacco, unspecified mental and behavioural disorder",
        "Mental and behavioural disorders due to use of cocaine, harmful use",
        "Mental and behavioural disorders due to use of cocaine, unspecified mental and behavioural disorder",
        "Intentional self-poisoning by and exposure to nonopioid analgesics, antipyretics and antirheumatics",
        "Accidental poisoning by and exposure to nonopioid analgesics, antipyretics and antirheumatics"
        ]
```

```python
combining = [
    "Drug poisonings (overdose) Unintentional (X40-X44)",
    "Drug poisonings (overdose) Suicide (X60-X64)",
    "Drug poisonings (overdose) Undetermined (Y10-Y14)",
    "All other drug-induced causes",
    "Drug poisonings (overdose) Homicide (X85)",
]
```

```python
vital_upd = vital[~vital["Cause of death"].isin(delete_cause)]
vital_upd
```
